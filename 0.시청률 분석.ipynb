```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# CSV 파일 불러오기
table1 = pd.read_csv('시청률/절친 토큐멘터리 4인용식탁.csv', parse_dates=['날짜'])
table2 = pd.read_csv('시청률/탐정들의 영업비밀.csv', parse_dates=['날짜'])
table3 = pd.read_csv('시청률/남의 나라 살아요-선 넘은 패밀리.csv', parse_dates=['날짜'])
table4 = pd.read_csv('시청률/새벽 2시의 신데렐라.csv', parse_dates=['날짜'])
table5 = pd.read_csv('시청률/성적을 부탁해 티처스.csv', parse_dates=['날짜'])
table6 = pd.read_csv('시청률/아빠는 꽃중년.csv', parse_dates=['날짜'])
table7 = pd.read_csv('시청률/오은영의 금쪽 상담소.csv', parse_dates=['날짜'])
table8 = pd.read_csv('시청률/요즘 남자 라이프 - 신랑수업.csv', parse_dates=['날짜'])
table9 = pd.read_csv('시청률/요즘 육아 금쪽같은 내 새끼.csv', parse_dates=['날짜'])
table10 = pd.read_csv('시청률/이제 만나러 갑니다.csv', parse_dates=['날짜'])

# 프로그램 이름과 데이터프레임을 딕셔너리에 저장
programs = {
    '절친 토큐멘터리 4인용식탁': table1,
    '탐정들의 영업비밀': table2,
    '남의 나라 살아요-선 넘은 패밀리': table3,
    '새벽 2시의 신데렐라': table4,
    '성적을 부탁해 티처스': table5,
    '아빠는 꽃중년': table6,
    '오은영의 금쪽 상담소': table7,
    '요즘 남자 라이프 - 신랑수업': table8,
    '요즘 육아 금쪽같은 내 새끼': table9,
    '이제 만나러 갑니다': table10
}

# 시청률 데이터를 숫자로 변환
for program_name, table in programs.items():
    # 시청률 열에서 % 기호를 제거하고 숫자로 변환
    table['시청률'] = table['시청률'].str.replace('%', '').astype(float)

# 기울기를 저장할 리스트
trend_slopes = []

# 각각의 프로그램에 대해 시청률 기울기 계산
for program_name, table in programs.items():
    # x 값은 날짜를 숫자로 변환 (일차)
    x = (table['날짜'] - table['날짜'].min()).dt.days
    y = table['시청률']
    
    # 선형 회귀 계산 (numpy polyfit을 이용해 1차 다항식을 구함)
    slope, _ = np.polyfit(x, y, 1)
    
    # 프로그램 이름과 기울기를 저장
    trend_slopes.append((program_name, slope))

# 기울기 데이터를 pandas DataFrame으로 변환
trend_df = pd.DataFrame(trend_slopes, columns=['프로그램 이름', '시청률 추세 기울기'])

# 기울기 기준으로 음수인 프로그램들만 선택
negative_trend_programs = trend_df[trend_df['시청률 추세 기울기'] < 0]

# 시청률 추세선을 그리기 위한 그래프 설정
plt.figure(figsize=(10, 6))

# 추세선을 그리는 함수
def plot_trendline(table, label):
    # x 값은 날짜를 숫자로 변환 (일차)
    x = (table['날짜'] - table['날짜'].min()).dt.days
    y = table['시청률']
    
    # 선형 회귀 계산 (numpy polyfit을 이용해 1차 다항식을 구함)
    slope, intercept = np.polyfit(x, y, 1)
    
    # 추세선 그리기
    plt.plot(table['날짜'], slope * x + intercept, linestyle='--', label=f'{label} 추세선')

# 추세가 음수인 프로그램들의 추세선 그리기
for program_name in negative_trend_programs['프로그램 이름']:
    table = programs[program_name]
    plot_trendline(table, program_name)

# 그래프 제목 및 축 레이블
plt.title('시청률 추세가 음수인 프로그램들의 추세선')
plt.xlabel('날짜')
plt.ylabel('시청률 (%)')

# 범례 추가
plt.legend()

# 날짜가 기울어지지 않게 x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.savefig('시청률_음수_추세선.png')
plt.show()
```
