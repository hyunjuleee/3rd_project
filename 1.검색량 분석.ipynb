```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# CSV 파일 불러오기
table1 = pd.read_csv('검색량/절친 토큐멘터리 4인용식탁.csv', parse_dates=['날짜'])
table2 = pd.read_csv('검색량/탐정들의 영업비밀.csv', parse_dates=['날짜'])
table3 = pd.read_csv('검색량/남의 나라 살아요-선 넘은 패밀리.csv', parse_dates=['날짜'])
table4 = pd.read_csv('검색량/새벽 2시의 신데렐라.csv', parse_dates=['날짜'])
table5 = pd.read_csv('검색량/성적을 부탁해 티처스.csv', parse_dates=['날짜'])
table6 = pd.read_csv('검색량/아빠는 꽃중년.csv', parse_dates=['날짜'])
table7 = pd.read_csv('검색량/오은영의 금쪽 상담소.csv', parse_dates=['날짜'])
table8 = pd.read_csv('검색량/요즘 남자 라이프 - 신랑수업.csv', parse_dates=['날짜'])
table9 = pd.read_csv('검색량/요즘 육아 금쪽같은 내 새끼.csv', parse_dates=['날짜'])
table10 = pd.read_csv('검색량/이제 만나러 갑니다.csv', parse_dates=['날짜'])
```

```python
# 시청률 비교를 위한 선 그래프 그리기
plt.figure(figsize=(14, 8))

# 각각의 테이블을 그래프에 추가
plt.plot(table1['날짜'], table1['검색량'], label='절친 토큐멘터리 4인용식탁', marker='o')
plt.plot(table2['날짜'], table2['검색량'], label='탐정들의 영업비밀', marker='o')
plt.plot(table3['날짜'], table3['검색량'], label='남의 나라 살아요-선 넘은 패밀리', marker='o')
plt.plot(table3['날짜'], table4['검색량'], label='새벽 2시의 신데렐라', marker='o')
plt.plot(table5['날짜'], table5['검색량'], label='성적을 부탁해 티처스', marker='o')
plt.plot(table6['날짜'], table6['검색량'], label='아빠는 꽃중년', marker='o')
plt.plot(table7['날짜'], table7['검색량'], label='오은영의 금쪽 상담소', marker='o')
plt.plot(table8['날짜'], table8['검색량'], label='요즘 남자 라이프 - 신랑수업', marker='o')
plt.plot(table9['날짜'], table9['검색량'], label='요즘 육아 금쪽같은 내 새끼', marker='o')
plt.plot(table10['날짜'], table10['검색량'], label='이제 만나러 갑니다', marker='o')

# 그래프 제목 및 축 레이블
plt.title('프로그램별 검색량 비교')
plt.xlabel('날짜')
plt.ylabel('검색량 (회)')

# 범례 추가
plt.legend()

# 날짜가 기울어지지 않게 x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.savefig('프라임타임10개프로그램_검색량')
plt.show()
```

```python
# 시청률 비교를 위한 선 그래프 그리기
plt.figure(figsize=(14, 8))

# 추세선을 그리는 함수
def plot_trendline(table, label):
    # x 값은 날짜를 숫자로 변환 (일차)
    x = (table['날짜'] - table['날짜'].min()).dt.days
    y = table['검색량']
    
    # 선형 회귀 계산 (numpy polyfit을 이용해 1차 다항식을 구함)
    slope, intercept = np.polyfit(x, y, 1)
    
    # 추세선 그리기
    plt.plot(table['날짜'], slope * x + intercept, linestyle='--', label=f'{label} 추세선')

# 각각의 테이블에 대해 추세선만 그리기
plot_trendline(table1, '절친 토큐멘터리 4인용식탁')
plot_trendline(table2, '탐정들의 영업비밀')
plot_trendline(table3, '남의 나라 살아요-선 넘은 패밀리')
plot_trendline(table5, '성적을 부탁해 티처스')
plot_trendline(table6, '아빠는 꽃중년')
plot_trendline(table7, '오은영의 금쪽 상담소')
plot_trendline(table8, '요즘 남자 라이프 - 신랑수업')
plot_trendline(table9, '요즘 육아 금쪽같은 내 새끼')
plot_trendline(table10, '이제 만나러 갑니다')

# 그래프 제목 및 축 레이블
plt.title('프로그램별 검색량 추세선 (새벽 2시의 신데렐라 제외)')
plt.xlabel('날짜')
plt.ylabel('검색량 (회)')

# 범례 추가
plt.legend()

# 날짜가 기울어지지 않게 x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.savefig('프라임타임10개프로그램_검색량_추세선')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# CSV 파일 불러오기
table1 = pd.read_csv('검색량/절친 토큐멘터리 4인용식탁.csv', parse_dates=['날짜'])
table2 = pd.read_csv('검색량/탐정들의 영업비밀.csv', parse_dates=['날짜'])
table3 = pd.read_csv('검색량/남의 나라 살아요-선 넘은 패밀리.csv', parse_dates=['날짜'])
table4 = pd.read_csv('검색량/새벽 2시의 신데렐라.csv', parse_dates=['날짜'])
table5 = pd.read_csv('검색량/성적을 부탁해 티처스.csv', parse_dates=['날짜'])
table6 = pd.read_csv('검색량/아빠는 꽃중년.csv', parse_dates=['날짜'])
table7 = pd.read_csv('검색량/오은영의 금쪽 상담소.csv', parse_dates=['날짜'])
table8 = pd.read_csv('검색량/요즘 남자 라이프 - 신랑수업.csv', parse_dates=['날짜'])
table9 = pd.read_csv('검색량/요즘 육아 금쪽같은 내 새끼.csv', parse_dates=['날짜'])
table10 = pd.read_csv('검색량/이제 만나러 갑니다.csv', parse_dates=['날짜'])

# 프로그램 이름과 데이터프레임을 딕셔너리에 저장
programs = {
    '절친 토큐멘터리 4인용식탁': table1,
    '탐정들의 영업비밀': table2,
    '남의 나라 살아요-선 넘은 패밀리': table3,
    '새벽 2시의 신데렐라': table4,
    '성적을 부탁해 티처스': table5,
    '아빠는 꽃중년': table6,
    '오은영의 금쪽 상담소': table7,
    '요즘 남자 라이프 - 신랑수업': table8,
    '요즘 육아 금쪽같은 내 새끼': table9,
    '이제 만나러 갑니다': table10
}

# 검색량 데이터를 숫자로 변환
for program_name, table in programs.items():
    if table['검색량'].dtype == 'object':  # 검색량이 문자열일 경우만 처리
        # 검색량 열이 문자열일 경우 숫자로 변환
        table['검색량'] = table['검색량'].str.replace(',', '').astype(float)

# 기울기를 저장할 리스트
trend_slopes = []

# 각각의 프로그램에 대해 검색량 기울기 계산
for program_name, table in programs.items():
    # x 값은 날짜를 숫자로 변환 (일차)
    x = (table['날짜'] - table['날짜'].min()).dt.days
    y = table['검색량']
    
    # 선형 회귀 계산 (numpy polyfit을 이용해 1차 다항식을 구함)
    slope, _ = np.polyfit(x, y, 1)
    
    # 프로그램 이름과 기울기를 저장
    trend_slopes.append((program_name, slope))

# 기울기 데이터를 pandas DataFrame으로 변환
trend_df = pd.DataFrame(trend_slopes, columns=['프로그램 이름', '검색량 추세 기울기'])

# 기울기 기준으로 양수인 프로그램들만 선택
positive_trend_programs = trend_df[trend_df['검색량 추세 기울기'] > 0]

# 검색량 추세선을 그리기 위한 그래프 설정
plt.figure(figsize=(10, 6))

# 추세선을 그리는 함수
def plot_trendline(table, label):
    # x 값은 날짜를 숫자로 변환 (일차)
    x = (table['날짜'] - table['날짜'].min()).dt.days
    y = table['검색량']
    
    # 선형 회귀 계산 (numpy polyfit을 이용해 1차 다항식을 구함)
    slope, intercept = np.polyfit(x, y, 1)
    
    # 추세선 그리기
    plt.plot(table['날짜'], slope * x + intercept, linestyle='--', label=f'{label} 추세선')

# 추세가 양수인 프로그램들의 추세선 그리기
for program_name in positive_trend_programs['프로그램 이름']:
    table = programs[program_name]
    plot_trendline(table, program_name)

# 그래프 제목 및 축 레이블
plt.title('검색량 추세가 양수인 프로그램들의 추세선')
plt.xlabel('날짜')
plt.ylabel('검색량')

# 범례 추가
plt.legend()

# 날짜가 기울어지지 않게 x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.savefig('검색량_양수_추세선.png')
plt.show()
```
